```python
import pickle
import numpy as np

# 读取 .pkl 文件
with open('../data/CVT/cvt_test_03.pkl', 'rb') as file:
    data = pickle.load(file)

# 计算每一列的最小值和最大值
min_vals = data.min(axis=0)
max_vals = data.max(axis=0)

# 对数据进行 Min-Max 标准化
data = (data - min_vals) / (max_vals - min_vals)

data = data[:50,:]
```

```python
import matplotlib.pyplot as plt
# 绘制每个维度的折线图
plt.figure(figsize=(10, 6))
x = np.arange(data.shape[0])  # 横轴索引

for i in range(data.shape[1]):
    plt.plot(x, data[:, i], label=f'Dimension {i+1}')

# 添加图例和标签
plt.title("Line Plots for Each Dimension")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True)

# 显示图形
plt.show()
```
